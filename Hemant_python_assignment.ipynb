# 3) Write a Python program to count the number 4 in a given list.

```python
def list_count_4(nums):
  count = 0  
  for num in nums:
    if num == 4:
      count = count + 1

  return count

print(list_count_4([1, 4, 6, 8, 4,9,4]))
```

# 8) Write a Python function that takes a list and returns a new list with unique elements of the first list.

```python
def unique_list(l):
  x = []
  for a in l:
    if a not in x:
      x.append(a)
  return x

print(unique_list([1,2,3,3,3,3,4,5])) 
```

```python
def dice():

    n = input( "Enter # of dice [3,6]: " )

    n = int( n )

    diceList = []
    for row in range( 0, n + 1 ):

        rowList = []
        for col in range( 0, n + 1 ):
            rowList.append( row * col )
        diceList.append( rowList )

    print( "\nYou have rolled: ", diceList )
dice()
```

```python
import random


def dice():
    n = input( "Enter # of dice [3,6]: " )
    n = int( n )
    diceList = []
    for row in range(0, n):
        diceList.append(random.randint(1, 6))
    print( "\nYou have rolled: ", diceList)


dice()
```

# 3) Using while loop accept numbers until sum of numbers is less than 100

```python
num = int(input("Enter a number: ")) 

if num < 0:
   print("Please enter a positive number")
else:
   sum = 0
   
   # use while loop to iterate until zero
   while(num > 0):
       sum += num
       num -= 1
   print("The result is", sum)
```
